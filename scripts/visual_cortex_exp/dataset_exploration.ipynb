```python
spike_train = dataset['resp_train']
spike_train_blank = dataset['resp_train_blk']
```

```python
spike_train_blank.shape
```

```python
#No of repeats = 20
#No of images  = 956
#No of neurons = 102
spike_train.shape
```

```python
latency = 50
end = spike_train.shape[3]
tmp = spike_train.copy()
tmp[:,:,:,:-latency] = spike_train[:,:,:,latency:]
tmp[:,:,:,end-latency:] = spike_train_blank[:,:,:,0:latency]
spike_train = tmp
```

```python
#spike count per stimulus presentation
resp = np.nansum(spike_train, 3)
```
